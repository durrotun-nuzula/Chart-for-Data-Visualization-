## Data visualization

Data visualization berkaitan dengan reporting activity to upper level. Most of them are business people and they want to know snap shots from the data. Pemilihan chart yang tepat dalam data visualization sangatlah penting agar upper level dapat memahaminya dengan cepat.

### 1. Bar CHART
Bar chart biasa digunakan untuk compare kuantitas atau frekuensi dari beberapa group

```python
df.columns
```

```python
daerah = df['Lokasi Pembelian'].value_counts()
daerah
```

```python
daerah.plot.barh()
plt.ylabel("Kota")
plt.xlabel("Frekuensi")
plt.title("Jumlah Penjulan berdasarkan kota")
```

```python
tipe=pd.crosstab(df["Tipe"], df['Jenis Pembayaran'])
```

```python
tipe
```

```python
tipe.plot.bar()
```

```python
Warna=pd.crosstab(df["Warna"], df['Jenis Kelamin'])
Warna.plot.bar()
```

### 2. Line CHART
line chart digunakan dalam data time series untuk mengetahui growth suatu produk.

```python
transaction = df['Tanggal Transaksi'].value_counts()
transaction
#gimana cara sort by time??? data tanggal sepertinya harus dirubah dulu.
```

```python
transaction.plot.line()
plt.title("Transaksi periode 24-29 April")
```

Urutkan dulu tanggal transaksinya. Kemudian dilihat trend perharinya.

```python
harga_jual = pd.pivot_table(df,index=["Tanggal Transaksi"],columns=["Jenis Pembayaran"],values=["Harga Jual"])
```

```python
harga_jual
```

```python
harga_jual.plot.line()
plt.title("Total Penjualan HP")
```

```python
Merek=pd.crosstab(df["Tanggal Transaksi"], df['Merek'])
```

```python
Merek.plot.line()
```

Trend penjualan HP pada semua merk mengalami penurunan pada akhir bulan April

```python
sns.swarmplot(y="Harga Jual", x="Jenis Kelamin" , data =df)
```

### 3. Pie CHART

```python
gender.plot.pie()
```

### 4. Boxplot dan histogram
Boxplot dan histogram sering digunakan untuk mengetahui persebaran dari data.

```python
df['Usia'].describe()
```

```python
plt.boxplot(df['Usia'])
plt.title("Persebaran Usia Pembeli HP")
```

Kebanyakan pembeli HP merupakan usia produktif

```python
plt.hist(df['Usia'], bins=20)
plt.title("Histogram Persebaran Usia Pembeli HP")
```
